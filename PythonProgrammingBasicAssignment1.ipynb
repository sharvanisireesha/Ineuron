```python
# Below are the the first five programs from the programming assignment1

#1.	Write a Python program to print "Hello Python"?
#2.	Write a Python program to do arithmetical operations addition and division.?
#3.	Write a Python program to find the area of a triangle?
#4.	Write a Python program to swap two variables?
#5.	Write a Python program to generate a random number?


import random


print("Hello Python")


def add(x, y):
    return x+y


def sub(x, y):
    return x-y


def mul(x, y):
    return x*y


def div(x, y):
    return x/y


def swap(x, y):
    _ = y
    y = x
    x = _
    return x, y


def generateRandomNumber(x, y):
    return random.randint(x, y)


def areaOfTriangle(base, height):
    return 1/2 * base * height


operation = input(
    '''please enter the operation to be performed.
    The choices are add,
                    sub,
                    mul,
                    div,
                    area,
                    swap,
                    randomval.  \n''')

if operation == 'add':
    num1 = int(input("Enter the num1 for Addition : "))
    num2 = int(input("Enter the num2 for Addition : "))

    value = add(num1, num2)
    print(f' The value of  {operation} operation is {value}')
elif operation == 'sub':
    num1 = int(input("Enter the num1 for subtraction : "))
    num2 = int(input("Enter the num2 for subtraction : "))

    value = sub(num1, num2)
    print(f' The value of  {operation} operation is {value}')
elif operation == 'mul':
    num1 = int(input("Enter the num1 for Multiplication : "))
    num2 = int(input("Enter the num2 for Multiplication : "))

    value = mul(num1, num2)
    print(f' The value of  {operation} operation is {value}')
elif operation == 'div':
    num1 = int(input("Enter the num1 for Division : "))
    num2 = int(input("Enter the num2 for Division : "))

    value = div(num1, num2)
    print(f' The value of  {operation} operation is {value}')
elif operation == 'swap':
    num1 = int(input("Enter the num1 for swapping : "))
    num2 = int(input("Enter the num2 for swapping : "))

    value1, value2 = swap(num1, num2)
    print(
        f' The value of  {operation} operation for  num1 is {value1} and num2 is {value2}')
elif operation == 'area':
    base = int(input("Enter the  value of base of the triangle : "))
    height = int(input("Enter the value for the height of the traingle : "))

    value = areaOfTriangle(base, height)
    print(f' The value of  {operation} operation is {value} units')
elif operation == 'randomval':
    num1 = int(input("Enter the Lower Boundary number : "))
    num2 = int(input("Enter the Upper boundary number : "))
    value = generateRandomNumber(num1, num2)
    print(f' The value of  {operation} operation is {value}')
else:
    print("please select a valid operation")
```
