```python
import mysql.connector as con
```

```python
mydb = con.connect(host = "localhost", user = "root" ,passwd = "root@123")
```

```python
cur = mydb.cursor()
```

```python
#2 creating the database cardataset
cur.execute(" create database cardataset")
```

```python
cur.execute("show databases")
```

```python
cur.fetchall()
```

```python
cur.execute("use cardataset")
```

```python
#3 creating the  car table by taking the  column values from  dataset description
cur.execute("create table car(buying varchar(15), maint varchar(15), doors varchar(15), persons varchar(15), lug_boot varchar(15), safety varchar(15), class varchar(15))")
```

```python
cur.execute("show tables")
```

```python
cur.fetchall()
```

```python
 #1 accessing the data from  accessed URL which is saved as car.data and #4 dumping all the data  from CSV into the car table
import csv

with open ("car.data","r") as f:
    cardata = csv.reader(f, delimiter = "\n")
    for i in cardata:
        strval = i[0]
        vallist = strval.split(',')
        
        cur.execute(f"insert into cardataset.car values(\"{str(vallist[0])}\",\"{str(vallist[1])}\", \"{str(vallist[2])}\", \"{str(vallist[3])}\", \"{str(vallist[4])}\", \"{str(vallist[5])}\",\"{str(vallist[6])}\")")
    mydb.commit()
        
        
```

```python
#5 Checking if data is available inside the table
cur.execute("select * from  cardataset.car")
```

```python
cur.fetchall()
```

```python
#6grouping the data with column 1 i.e buying and counting the occurences based on col1
cur.execute("select  count(buying), buying from cardataset.car group by buying" )
```

```python
cur.fetchall()
```

```python
#7Filtering the records where the value of col3 doors is 4
cur.execute("select  * from cardataset.car where doors  = 4")
```

```python
cur.fetchall()
```

```python
# 8 Update the col3 which is doors value to 8 whenever doors = 2
cur.execute("update cardataset.car set doors = '8' where doors ='2'")
```

```python
 #Checking doors changed to 8 where doors =2
cur.execute("select * from  cardataset.car")
```
